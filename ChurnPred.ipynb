# Loading model

```python
df = pd.read_csv('IT_customer_churn.csv')
```

```python
df.head()
```

```python
df.dtypes
```

# Playing with DataFrame

```python
def WhatsGoingOn(df):
  for par in df.columns:
    print(par, '=>', df[par].unique())

WhatsGoingOn(df)
```

```python
# Numerication

df.replace({'No phone service':0}, inplace=True)
df.replace({'No internet service':0}, inplace=True)
df.replace({'Female':0, 'Male':1}, inplace=True)
df.replace({'No':0, 'Yes':1}, inplace=True)
```

```python
# Drop the rows which its TotalCharges is ' '
df = df[df['TotalCharges'] != ' ']
df.reset_index()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges)
```

```python
WhatsGoingOn(df)
```

```python
# Numerication-2
df['Contract'] = df['Contract'].replace({'Month-to-month':1, 'One year':12, 'Two year':24})
df['InternetService'] = df['InternetService'].replace({'DSL':1, 'Fiber optic':2})
```

```python
# get-dummies
pd.get_dummies(df, ['PaymentMethod'])
```

```python
df.dtypes
```

```python
df.dtypes
```

#Splitting the data

```python
X = df.drop(['Churn', 'PaymentMethod'], axis='columns')
Y = df.Churn
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Training the model
Logistic Regression

 RMSE

```python
model1 = LogisticRegression()
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
res = classification_report(y_test, y_pred)
print(res)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
x_sm, y_sm = smote.fit_resample(X, Y)
y_sm.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sm, y_sm)
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
res = classification_report(y_test, y_pred)
print(res)
```

```python
model2 = DecisionTreeClassifier()
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
res = classification_report(y_test, y_pred)
print(res)
```

```python
model3 = RandomForestClassifier()
model3.fit(x_train, y_train)
y_pred = model3.predict(x_test)
res = classification_report(y_test, y_pred)
print(res)
```

```python
model4 = XGBClassifier()
model4.fit(x_train, y_train)
y_pred = model4.predict(x_test)
res = classification_report(y_test, y_pred)
print(res)
```
